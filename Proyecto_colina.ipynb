```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import PolarAxes
import seaborn as sns
```

```python
names =['fecha','pm10', 'pm2_5', 'velocidad', 'precipitacion', 'direccion', 'o3', 'co', 'no', 'no2', 'nox', 'so2', 'temperatura', 'radiacion', 'hr']
colina = pd.read_csv('https://raw.githubusercontent.com/PedroCardenasB/CSA_USA/main/Colina_2023.csv', sep=';', parse_dates=True)
colina.columns = names
colina.head()
```

```python
data2 = colina [['fecha','pm10', 'pm2_5', 'velocidad', 'precipitacion', 'direccion', 'o3', 'co', 'no', 'no2', 'nox', 'so2', 'temperatura', 'radiacion', 'hr']]
data2 ['pm10'] = data2 ['pm10'].map(lambda x: str(x).replace(',', '.')).copy()
data2 ['pm2_5'] = data2 ['pm2_5'].map(lambda x: str(x).replace(',', '.')).copy()
data2 ['velocidad'] = data2 ['velocidad'].map(lambda x: str(x).replace(',', '.')).copy()
data2 ['precipitacion'] = data2 ['precipitacion'].map(lambda x: str(x).replace(',', '.')).copy()
data2 ['direccion'] = data2 ['direccion'].map(lambda x: str(x).replace(',', '.')).copy()
data2 ['o3'] = data2 ['o3'].map(lambda x: str(x).replace(',', '.')).copy()
data2 ['co'] = data2 ['co'].map(lambda x: str(x).replace(',', '.')).copy()
data2 ['no'] = data2 ['no'].map(lambda x: str(x).replace(',', '.')).copy()
data2 ['no2'] = data2 ['no2'].map(lambda x: str(x).replace(',', '.')).copy()
data2 ['nox'] = data2 ['nox'].map(lambda x: str(x).replace(',', '.')).copy()
data2 ['so2'] = data2 ['so2'].map(lambda x: str(x).replace(',', '.')).copy()
data2 ['temperatura'] = data2 ['temperatura'].map(lambda x: str(x).replace(',', '.')).copy()
data2 ['radiacion'] = data2 ['radiacion'].map(lambda x: str(x).replace(',', '.')).copy()
data2 ['hr'] = data2 ['hr'].map(lambda x: str(x).replace(',', '.')).copy()

data2['pm10']= data2['pm10'].astype(float)
data2['pm2_5']=data2['pm2_5'].astype(float)
data2['velocidad']=data2['velocidad'].astype(float)
data2['precipitacion']=data2['precipitacion'].astype(float)
data2['direccion']=data2['direccion'].astype(float)
data2['o3']=data2['o3'].astype(float)
data2['co']=data2['co'].astype(float)
data2['no']=data2['no'].astype(float)
data2['no2']=data2['no2'].astype(float)
data2['nox']=data2['nox'].astype(float)
data2['so2']=data2['so2'].astype(float)
data2['temperatura']=data2['temperatura'].astype(float)
data2['radiacion']=data2['radiacion'].astype(float)
data2['hr']=data2['hr'].astype(float)

data2.head()
```

```python
estadisticas = data2['velocidad'].describe()

valor_minimo = estadisticas['min']
valor_maximo = estadisticas['max']
promedio = estadisticas ['mean']

print("Valor mínimo:", valor_minimo)
print("Valor máximo:",valor_maximo)
print("Promedio:", promedio)
```

```python
def calcular_promedio_contaminante(contaminante):
  """Calcula el promedio de un contaminante específico.

  Args:
    contaminante: El nombre del contaminante ('pm10', 'pm2_5', 'no2', etc.).

  Returns:
    El promedio del contaminante.
  """
  promedio = data2[contaminante].mean()
  print(f"Promedio de {contaminante}: {promedio}")

promedio_pm10 = data2['pm10'].mean()
promedio_pm2_5 = data2['pm2_5'].mean()
promedio_no2 = data2['no2'].mean()

calcular_promedio_contaminante('pm10')
calcular_promedio_contaminante('pm2_5')
calcular_promedio_contaminante('no2')
```

```python
promedios_contaminantes = {
    'PM10': promedio_pm10,
    'PM2.5': promedio_pm2_5,
    'no2': promedio_no2
}

plt.figure(figsize=(10, 6))
plt.bar(promedios_contaminantes.keys(), promedios_contaminantes.values(), color=['skyblue', 'lightgreen', 'salmon'])

plt.title('Concentración Promedio de Contaminantes', fontsize=16)
plt.xlabel('Contaminantes', fontsize=14)
plt.ylabel('Concentración Promedio', fontsize=14)

for i, value in enumerate(promedios_contaminantes.values()):
    plt.text(i, value + 0.05, f'{value:.2f}', ha='center', fontsize=12)

plt.show()
```

```python
data2['fecha'] = pd.to_datetime(data2['fecha'], errors='coerce')
def determinar_estacion(fecha):
    month = fecha.month
    if month in [12, 1, 2]:
        return 'Invierno'
    elif month in [3, 4, 5]:
        return 'Primavera'
    elif month in [6, 7, 8]:
        return 'Verano'
    elif month in [9, 10, 11]:
        return 'Otoño'
data2['estacion'] = data2['fecha'].apply(determinar_estacion)
promedios_estacionales = data2.groupby('estacion')[['pm10', 'pm2_5', 'co']].mean()
print(promedios_estacionales)
promedios_estacionales.plot(kind='bar', figsize=(10, 6))
plt.title('Promedio de Concentraciones por Estación del año')
plt.ylabel('Concentración Promedio ')
plt.xlabel('Estación del año')
plt.xticks(rotation=45)
plt.show()
```

```python
# Asegurarse de que la columna 'fecha' esté en formato datetime
data2['fecha'] = pd.to_datetime(data2['fecha'], format='%d/%m/%Y %H:%M')

# Extraer el mes de la fecha
data2['mes'] = data2['fecha'].dt.month

# Graficar la rosa de los vientos por cada mes
fig, axes = plt.subplots(3, 4, figsize=(20, 15), subplot_kw=dict(projection='windrose'))
fig.suptitle('Rosas de los Vientos por Mes - Colina 2023', fontsize=20, y=0.92)

meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']

for i, ax in enumerate(axes.flatten()):
    # Filtrar los datos por mes
    data_mes = data2[data2['mes'] == i + 1]

    # Crear la rosa de los vientos
    ax.bar(data_mes['direccion'], data_mes['velocidad'], normed=True, opening=0.8, edgecolor='white')
    ax.set_title(meses[i], fontsize=16, pad=20)
    ax.set_legend()

plt.tight_layout()
plt.show()
```
